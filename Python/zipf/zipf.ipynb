### Zipf's Law

_the rank-frequency distribution is an inverse relation_

Demonstrated here by plotting word rank to word frequency for James Joyce's _Ulysses_

```python
# data is list of values
filename = "ulysses.txt" # from Project Gutenberg
data = open(filename,errors='ignore').read().split()
```

```python
count = {}
for _ in data:
    if _ in count.keys(): count[_]+=1
    else: count[_]=1
count = dict(sorted(count.items(),key=lambda _:_[1],reverse=True)) # dictionary of frequency
```

```python
freq = tuple(count.values())
rank = tuple(range(len(freq)))
```

```python
import matplotlib.pyplot as plt
plt.plot(rank,freq,'o')
plt.plot([rank[1],rank[-1]],[freq[1],freq[-1]])
plt.xscale("log")
plt.yscale("log")
plt.xlabel('rank')
plt.ylabel('frequency')
plt.show()
```
